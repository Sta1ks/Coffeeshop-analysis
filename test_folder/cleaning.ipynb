```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df= pd.read_excel("Coffee Shop Sales.xlsx")
df
```

```python
df = df.drop(["transaction_id", 'transaction_time'], axis=1)
```

```python
df["month"] = df["transaction_date"].dt.month_name()
df["day"] = df["transaction_date"].dt.day_name()
```

```python
df
```

```python
df = df.drop("transaction_date", axis=1)
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df
```

```python
df['store_location'].unique()
```

```python
enc = OneHotEncoder(sparse_output=False).set_output(transform="pandas")
enc_stores = enc.fit_transform(df[["store_location"]])
enc_stores
```

```python
df = pd.concat([df, enc_stores], axis = 1).drop(["store_location"], axis = 1)
```

```python
df = df.drop("store_location_Lower Manhattan", axis = 1)
```

```python
df.columns
```

```python
df.head(2)
```

```python
week_enc = enc.fit_transform(df[['month']])
week_enc
```

```python
df = pd.concat([df, week_enc], axis= 1).drop(['month'], axis  = 1)
```

```python
df = df.drop('month_May', axis = 1)
df.head(2)
```

```python
day_enc = enc.fit_transform(df[['day']])
df = pd.concat([df, day_enc], axis= 1).drop(['day'], axis  = 1)
df = df.drop('day_Sunday', axis = 1)
df.head(2)
```
